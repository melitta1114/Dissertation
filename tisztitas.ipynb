```python
import pandas as pd
import os
import chardet
```

```python
folder_path = r"D:\MESTERI\4. FELEV\Datasetek\data"
```

```python
results = []

# Bejárjuk az összes fájlt a mappában
for filename in os.listdir(folder_path):
    if filename.lower().endswith((".csv", ".txt")):  # CSV és TXT fájlok
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "rb") as f:
            raw = f.read()
        det = chardet.detect(raw)
        results.append({
            "filename": filename,
            "encoding": det["encoding"]
        })
```

```python
# Eredmények mentése egy CSV fájlba
df = pd.DataFrame(results)
df.to_csv(os.path.join(folder_path, "file_encoding_results.csv"), index=False, encoding="utf-8")
```
